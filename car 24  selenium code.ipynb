```python
Brand=[]
Price=[]
History=[]
Owner=[]
Kilometers=[]
Fuel_type=[]
Last_service=[]
Transmisson=[]
Insurance=[]
Registration=[]
Year_of_manufacturing=[]
```

```python
# 500 time we scroll down by 10000 in order to generate more cars on the website
for _ in range(100):
    driver.execute_script("window.scrollBy(0,100)")
```

```python
# scrape all product urls
product_urls = []

url=driver.find_elements_by_xpath('//div[@class="col-4"]/a')       #scraping urls
for i in url:
    product_urls.append(i.get_attribute("href"))                        #appending the urls in product_urls list
    time.sleep(2)


product_urls  
```

```python


for url in product_urls:
    
    driver.get(url)
    time.sleep(1)
    
     #Extracting Price
    try:
        price = driver.find_element_by_xpath('//li[@class="_2DkPb"]/span[2]')     
        Price.append(price.text)  
    except NoSuchElementException:
        Price.append('-')  
    
    # Extracting Brand from xpath 
    try:
        brand = driver.find_element_by_xpath('//div[@class="_10kZS"]/h2')      
        Brand.append(brand.text)  
    except NoSuchElementException:
        Brand.append('-')    


    time.sleep(1)    
    # Extracting history    
    try: 
        history = driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[1]/strong')  
        History.append(history.text)
    except NoSuchElementException:
        History.append('-')
        
    time.sleep(1)
    # Extracting owner
    try:
        owner = driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[2]/strong')     
        Owner.append(owner.text)
    except NoSuchElementException:
        Owner.append('-')


    time.sleep(1)
    # Extracting 
    try:
        kilometer = driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[3]/strong')  
        Kilometers.append(kilometer.text)
    except NoSuchElementException:
        Kilometers.append('-') 


    time.sleep(1)
    # Extracting fule type from xpath    
    try:
        fuel = driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[4]/strong') 
        Fuel_type.append(fuel.text)
    except NoSuchElementException:
        Fuel_type.append('-')
        
        
    time.sleep(1)
    # Extracting last service
    try:
        last_service= driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[5]/strong') 
        Last_service.append(last_service.text)
    except NoSuchElementException:
        Last_service.append('-')


    time.sleep(1)
    # Extracting transmission
    try:
        transmission= driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[6]/strong') 
        Transmisson.append(transmission.text)
    except NoSuchElementException:
        Transmisson.append('-')


    time.sleep(1)
    # Extracting registration
    try:
        registration= driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[7]/strong')  
        Registration.append(registration.text)
    except NoSuchElementException:
         Registration.append('-')


    time.sleep(1)
    # Extracting insurance       
    try:
        insurance= driver.find_element_by_xpath('//ul[@class="_2KO13"]/li[8]/strong')  
        Insurance.append(insurance.text)
    except NoSuchElementException:
        Insurance.append('-')
 
 # Extracting year of manufacturing     
    try:
        year= driver.find_element_by_xpath('//ul[@class="_2geSF"]/li[9]/strong')  
        Year_of_manufacturing.append(year.text)
    except NoSuchElementException:
        Year_of_manufacturing.append('-')
```
